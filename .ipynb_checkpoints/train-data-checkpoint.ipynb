```python
import scipy.io
from PIL import Image
import os
# Caminho para o arquivo de anota√ß√µes
data = os.path.join(target_directory, "stanford-cars-dataset", "cars_annos.mat")

# Carregar anota√ß√µes e nomes das classes
annotations = scipy.io.loadmat(data)['annotations'][0]
class_names = scipy.io.loadmat(data)['class_names'][0]

# Fun√ß√£o para converter anota√ß√µes para o formato YOLO
def convert_to_yolo_format(annotation, img_width, img_height):
    # Extrair as coordenadas da bounding box
    x_min = annotation['bbox_x1'][0][0]
    y_min = annotation['bbox_y1'][0][0]
    x_max = annotation['bbox_x2'][0][0]
    y_max = annotation['bbox_y2'][0][0]

    # Verifica se os valores da bounding box est√£o corretos
    if x_max < x_min or y_max < y_min:
        print(f"Erro na anota√ß√£o: x_min ({x_min}) > x_max ({x_max}) ou y_min ({y_min}) > y_max ({y_max})")
        return None  # Retorna None para indicar erro

    if img_width == 0 or img_height == 0:
        print(f"Erro: Imagem com dimens√£o inv√°lida ({img_width}x{img_height})")
        return None  # Retorna None para evitar erro

    # Calcular coordenadas YOLO (normalizadas)
    x_center = ((x_min + x_max) / 2) / img_width
    y_center = ((y_min + y_max) / 2) / img_height
    width = (x_max - x_min) / img_width
    height = (y_max - y_min) / img_height

    # Garantir que as coordenadas estejam dentro do intervalo [0, 1]
    x_center = min(1, max(0, x_center))
    y_center = min(1, max(0, y_center))
    width = min(1, max(0, width))
    height = min(1, max(0, height))

    return x_center, y_center, width, height
    
class_names = [name[0] for name in class_names]
print(class_names)
```

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt


def visualize_dataset(image_dir, label_dir, num_samples=3):
    image_files = list(Path(image_dir).glob('*.jpg'))[:num_samples]
    for image_file in image_files:
        img = Image.open(image_file)
        label_file = Path(label_dir) / (image_file.stem + '.txt')
        
        print(f"Image: {image_file.name}")
        
        if label_file.exists():
            labels = label_file.read_text().splitlines()
            for label in labels:
                # A label no formato YOLO √© composta por: class_id center_x center_y width height
                label_parts = label.split()
                class_id = int(label_parts[0])
                class_name = class_names[class_id] if class_id < len(class_names) else 'Unknown'
                print(f"Class ID: {class_id}, Class Name: {class_name}")
        else:
            print("No labels found")
        
        plt.imshow(img)
        plt.axis('off')
        plt.show()
```

```python
# Diret√≥rio das imagens de treino
image_dir = os.path.join(target_directory, 'images/train')
label_dir = os.path.join(target_directory, 'labels/train')  
os.makedirs(label_dir, exist_ok=True)
```

```python
for i in range(len(annotations)):
    
    image_name = annotations[i][0][0]
    class_id = annotations[i]['class'][0][0] -1  # As classes come√ßam em 0


    image_name_parts = image_name.split('/')


    while len(image_name_parts[-1].split('.')[0]) > 5:
        image_name_parts[-1] = image_name_parts[-1][1:]


    image_path = os.path.join(image_dir, image_name_parts[-1])


    if not os.path.exists(image_path):
        continue


    with Image.open(image_path) as img:
        img_width, img_height = img.size


    x_center, y_center, width, height = convert_to_yolo_format(annotations[i], img_width, img_height)


    label_file = os.path.join(label_dir, os.path.splitext(image_name_parts[-1])[0] + '.txt')
    with open(label_file, 'w') as f:
        f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```python
# Diret√≥rio das imagens de valida√ß√£o
val_image_dir = os.path.join(target_directory, 'images/val')
val_label_dir = os.path.join(target_directory, 'labels/val') 
os.makedirs(val_label_dir, exist_ok=True)
```

```python
for i in range(len(annotations)):  
    
    image_name = annotations[i][0][0]
    class_id = annotations[i]['class'][0][0] - 1  # As classes come√ßam em 0


    image_name_parts = image_name.split('/')


    while len(image_name_parts[-1].split('.')[0]) > 5:
        image_name_parts[-1] = image_name_parts[-1][1:]


    image_path = os.path.join(val_image_dir, image_name_parts[-1])


    if not os.path.exists(image_path):
        continue


    with Image.open(image_path) as img:
        img_width, img_height = img.size


    x_center, y_center, width, height = convert_to_yolo_format(annotations[i], img_width, img_height)


    label_file = os.path.join(val_label_dir, os.path.splitext(image_name_parts[-1])[0] + '.txt')
    with open(label_file, 'w') as f:
        f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```python
visualize_dataset(image_dir, label_dir)
```

```python
import yaml
import os
import scipy.io

# Caminho para o arquivo de anota√ß√µes
data_path = os.path.join(target_directory, "stanford-cars-dataset", "cars_annos.mat")

# Carregar anota√ß√µes e nomes das classes
annotations = scipy.io.loadmat(data_path)['annotations'][0]
class_names = [str(name[0]) for name in scipy.io.loadmat(data_path)['class_names'][0]]  # Extract car model names

# Configura√ß√£o do YAML para treinamento
data = {
    'train': os.path.join(target_directory, 'images/train'),
    'val': os.path.join(target_directory, 'images/val'),
    'nc': len(class_names),  # Total number of car classes
    'names': class_names,  # Car model names
    'train_labels': os.path.join(target_directory, 'labels/train'),
    'val_labels': os.path.join(target_directory, 'labels/val')
}

# Salvar arquivo YAML
yaml_path = os.path.join(target_directory, 'stanford_cars.yaml')
with open(yaml_path, 'w') as f:
    yaml.dump(data, f, default_flow_style=False)

print(f"Arquivo YAML criado em: {yaml_path}")
```

```python
import cv2

from ultralytics import YOLO
import torch
import torchvision

# Verificar disponibilidade da GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")


# Carregar modelo YOLOv8 pr√©-treinado
model = YOLO('yolov8n.pt')  # YOLOv8 Nano
```

```python
# Fine-Tuning do YOLOv8
epochs = 10

model.train(
    data=yaml_path,         # Caminho para o arquivo YAML
    epochs=epochs,              # N√∫mero de √©pocas
    imgsz=640,              # Tamanho da imagem
    batch=16,               # Tamanho do batch
    workers=4,              # N√∫mero de workers
    device=device,          # Usar GPU se dispon√≠vel
    name="yolo_cars_finetune",  # Nome do experimento
    project="stanford_cars",    # Nome do projeto
    exist_ok=True           # Sobrescrever pastas existentes
)
```
